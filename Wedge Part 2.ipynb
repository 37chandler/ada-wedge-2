## Part 2: Build a list of owners

```python
query = ("select distinct card_no,"
"from `wedge-project-290522.big_wedge_1.transArchive*`"
"Where card_no != 3")

#execute the query using `client.query`

query_job = client.query(query,location = "US",)
```

```python
owners = []

for idx, row in enumerate(query_job) :
    card_no = row[0] #since row in GBQ is at uple, you need to place [] and 

    owners.append(card_no)
```

## Part 3:	Takes a sample of the owners. 

```python
#some sort of random choices code, "k="" then tell it how many you want
random.seed(20201015)  #start the random number generator at the same spot everytime
owner_sample = random.choices(owners,k=350)
```

## Part 4: Extracts all records associated with those owners and writes them to a local text file. 

```python
query = """select * from `wedge-project-290522.big_wedge_1.transArchive*` Where card_no in ( """

query = query + ",".join([str(owner) for owner in owner_sample]) + ")"
```

```python
query_job = client.query(query,location = "US",)
```
